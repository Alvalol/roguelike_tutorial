```python
import tcod
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def plot_noise(dimensions=1, hurst=0.5, lacunarity=2, octaves=4):
    noise = tcod.noise.Noise(
        dimensions=dimensions,
        algorithm=tcod.NOISE_SIMPLEX,
        implementation=tcod.noise.FBM,
        hurst=hurst,
        lacunarity=lacunarity,
        octaves=octaves,
        # seed=tcod.random.Random(1, seed=16),
        )

    # Create a 5x5 open multi-dimensional mesh-grid.
    ogrid = np.arange(0, 10, 0.03, dtype=np.float32)

    # Scale the grid.
    #ogrid[0] *= 0.25
    #ogrid[1] *= 0.25

    # Return the sampled noise from this grid of points.
    samples = noise.sample_ogrid([ogrid])
    
    # Try to to smooth samples to zerp at the beginning and end
    # samples[:10] = samples[:10] * np.arange(0, 1, 0.1)
    # samples[-10:] = samples[-10:] * np.arange(1, 0, -0.1)

    plt.figure(figsize=(4,3))
    plt.plot(ogrid,samples)
    plt.show()
```

```python
for _ in range(10):
    plot_noise(dimensions=1, hurst=0.5, lacunarity=2, octaves=4)
```
